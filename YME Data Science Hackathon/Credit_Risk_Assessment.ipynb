# We explore the age data to have an idea of how we can clean it. 


*   We can see that the ages that are not available are listed as 0.0.
*   We also check if there are any abnormalities. In our case, we have two seemingly valid ages, but one has to be above 18 to obtain a personal credit card.



```python
print("Minimum age: " + str(min(df["age"])))
print("Maximum age: " + str(max(df["age"])))
df[(df["age"] > 10) & (df["age"] < 18)].head(10)
```

```python
age_df = df[df["age"] > 18]
accepted_ages = age_df[(age_df["accepted"] == 1)]
non_acp_ages = age_df[(age_df["accepted"] != 1)]
print("Mean age of accepted applicants: %.3f" % np.mean(accepted_ages["age"]))
print("Mean age of un-accepted applicants: %.3f" % np.mean(non_acp_ages["age"]))
print("Mean age: %.3f" % np.mean(age_df["age"]))
```
